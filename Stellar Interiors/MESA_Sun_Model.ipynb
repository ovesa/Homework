MESA was used to create an initial model of the Sun (1 solar mass). That model was plotted onto an HR Diagram. The Sun's current location is also marked. Another model was created by changing one factor to try to get closer near observed solar values.

```python
import mesa_reader as mr
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rc
plt.rc('text', usetex=True)
```

```python
# upload relevant history.data files of both runs
h = mr.MesaData('LOGS/history2.data')
h2 = mr.MesaData('LOGS/history.data')
#h2 = mr.MesaData('LOGS/history.data')
```

```python
# Sun's information
Tsol = 5770 #K
Lsol = 1 # solar luminosity
```

```python
np.log10(Tsol)
```

```python
print(plt.style.available)
plt.style.use('classic')
from pylab import *
rcParams['legend.numpoints'] = 1
```

```python

plt.figure(figsize=(10,10),facecolor="white")
plt.plot(h.c, h.log_L, label ='Run 1')
plt.plot(h2.log_Teff, h2.log_L, linestyle='--', label ='Run 2',c='k',linewidth=2)

plt.plot(np.log10(Tsol),np.log10(Lsol), c='orange', marker='*', label='Sun',linewidth=0,markersize=17)
plt.xlabel(r'$\log (T_{eff})$', fontweight='bold', fontsize=18)
plt.ylabel(r'$\log \left( L/L_\odot \right)$', fontweight='bold', fontsize=18)
plt.title('The Journey of One Sun', fontweight='bold', fontsize=14)
plt.legend(fancybox=True,framealpha=0.5,loc='best')
plt.xlim(right=3.8)
plt.ylim(bottom=-0.5,top=2.8)
plt.gca().invert_xaxis()
plt.savefig('Thejourneyofonesun.png')
plt.show()
```
